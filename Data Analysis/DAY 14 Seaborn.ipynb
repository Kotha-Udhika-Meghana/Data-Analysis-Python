## Palette
- color_palette()
- palplot()
- it provides diff shades of colors.
- it has 3 types
    + *Qualitative*
        - deep
        - muted
        - bright
        - pastel
        - dark
        - colorblind
    + *Sequential*
    + *Diverging*

```python
help(sns.color_palette)
```

## Qualitative

```python
sns.palplot(sns.color_palette())
```

```python
sns.palplot(sns.color_palette('deep'))
```

```python
sns.palplot(sns.color_palette('bright'))
```

```python
sns.palplot(sns.color_palette('muted'))
```

```python
sns.palplot(sns.color_palette('pastel'))
```

```python
sns.palplot(sns.color_palette('colorblind'))
```

## Sequential

```python
## Ligh - brightest color shades - low to high
sns.palplot(sns.color_palette('Reds'))
```

```python
## _r represents reverse order - high to low
sns.palplot(sns.color_palette('Greens_r'))
```

```python
## _r represents reverse order - high to low
sns.palplot(sns.color_palette('Greys_r'))
```

```python
## Ligh - brightest color shades - low to high
sns.palplot(sns.color_palette('Blues',n_colors = 10))
```

## Diverging

```python
sns.palplot(sns.color_palette('BrBG', 10))
```

```python
sns.palplot(sns.dark_palette('Red'))
```

```python
sns.palplot(sns.light_palette('Red'))
```

```python
##  convert dark to light
sns.palplot(sns.light_palette('Red',reverse=True))
```

## Line plot using Seaborn

```python
x1 = [1,2,3,4,5]
x2 = [36.6,37,37.7,56,44]
## to make these two x1,x2 into dataframe
df1 = pd.DataFrame({'days':x1,'temp':x2})
df1
```

```python
sns.lineplot(x="days",y="temp",data=df1)
plt.show()
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
tips.shape
```

```python
sns.lineplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lineplot(x='size',y='total_bill',data=tips)
```

```python
help(sns.lineplot)
```

```python
sns.lineplot(x='size',y='total_bill',data=tips,hue="sex",palette='hot')
```

```python
sns.lineplot(x='size',y='total_bill',data=tips,hue="sex",palette='hot',legend="brief",markers=["o","^"])
```

```python
plt.figure(figsize=(16,9))
sns.set(style="whitegrid")
sns.lineplot(x='size',y='total_bill',data=tips,hue="sex",palette='hot',legend="brief",markers=["o","^"])
```

```python
plt.figure(figsize=(16,9))
sns.set(style="whitegrid")
sns.lineplot(x='size',y='total_bill',data=tips,hue="day",palette='hot',legend="brief",markers=True)
```

## Distribution Plots
- Univariate Distribution Plot.
- Bivariate Distribution Plot.

```python
## Univariate
sns.distplot(tips['size'])
```

```python
sns.distplot(tips['tip'])
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['tip'],bins=20)
```

```python
help(sns.distplot)
```

```python
sns.distplot(tips['tip'],hist=False)
```

```python
sns.distplot(tips['tip'],hist=False,kde=False,rug=True)
```

```python
sns.distplot(tips['tip'],color='k',vertical=True,label='tip')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16,10))
sns.distplot(tips['tip'],color='k',vertical=True,label='tip')
plt.legend()
sns.set_context('paper')
plt.show()
```

```python
sns.distplot(tips['tip'],hist_kws={"color":'m',"edgecolor":'k',"alpha":0.9,"linewidth":2,'linestyle':'--'})
plt.legend()
plt.show()
```
