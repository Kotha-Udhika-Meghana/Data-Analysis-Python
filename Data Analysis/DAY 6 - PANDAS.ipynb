### Merge

```python
help(pd.merge)
```

```python
F1 = pd.DataFrame({"Employee":pd.Series(["M","D","L","N","V"],index = [1,2,3,4,5]),
     "Year":pd.Series([2000,2001,2002,2003,2004],index = [1,2,3,4,5])})
print(F1)
```

```python
F2 = pd.DataFrame({"Employee":pd.Series(["M","D","L","P"],index = [1,2,3,4]),
    "Department":pd.Series(["hr","op","ma","s"],index = [1,2,3,4])})
print(F2)
```

#### Merging above two frames

```python
pd.merge(F1,F2)   ## Based on emp names merges is done
```

```python
pd.merge(F1,F2,how="left") ## Left join
```

```python
pd.merge(F1,F2,how="right") ## Right Join
```

```python
dF1 = pd.DataFrame({'lkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [1, 2, 3, 5]})
dF2 = pd.DataFrame({'rkey': ['foo', 'bar', 'baz', 'foo'], 
                    'value': [5, 6, 7, 8]})
print(dF1)
print(dF2)
```

```python
dF1.merge(dF2)
```

```python
dF1.merge(dF2, how ="left")
```

```python
dF1.merge(dF2, how ="right")
```
