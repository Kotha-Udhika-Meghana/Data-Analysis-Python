```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
cancer_dataset = load_breast_cancer()
cancer_df = pd.DataFrame(cancer_dataset.data,columns=cancer_dataset.feature_names)
cancer_df["target"] = cancer_dataset.target
```

```python
cancer_df.head()
```

```python
sns.pairplot(cancer_df,vars=['mean radius','mean texture','mean perimeter','mean area','mean smoothness'])
```

```python
sns.pairplot(cancer_df,vars=['mean radius','mean texture','mean perimeter','mean area','mean smoothness'],hue='target')
```

```python
sns.pairplot(cancer_df,vars=['mean radius','mean texture','mean perimeter','mean area','mean smoothness'],hue='target',hue_order=[1.0,0.0])
```

```python
sns.pairplot(cancer_df,vars=['mean radius','mean texture','mean perimeter',
'mean area','mean smoothness'],hue='target',palette="Dark2")
```

```python
sns.pairplot(cancer_df,x_vars=['mean radius','mean texture'],y_vars=['mean perimeter'],hue='target')
```

```python
sns.pairplot(cancer_df,x_vars=['mean radius','mean texture'],y_vars=['mean perimeter','mean area'],hue='target')
```

```python
sns.pairplot(cancer_df,x_vars=['mean radius','mean texture'],
        y_vars=['mean perimeter','mean area'],hue='target',markers=['o','^'])
```
