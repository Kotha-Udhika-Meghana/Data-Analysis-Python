### Scaling Techniques:
- it is used to compress the value btw a fixed range (0 to 1 or mean to std or -ve to +ve )
    * Standard Scaler
    * Min Max Scaler
    * Robust Scaler  
    * Normalizer

### 1.Standard Scaler
- By applying it, the dataset will be normally distributed.
- It scales the data point btw mean to unit variance.
     + **Formula**
         - Z = (XI - Mean(X)) / std(X)

### 2. MinMax  Scaler
- It scales the data points btw 0 to 1
    * **Formula**
        + [Xi - min(X)] / [max(X) - min(X)]

```python
Df.head(3)
```

```python
print(Df['TV'].min())
print(Df['TV'].max())
```

```python
(230.1 - 0.7) / (296.4 - 0.7)
```

```python
## Import the minmax mdoel first
from sklearn.preprocessing import MinMaxScaler
```

```python
## Call the minmax model and initialise it
ms = MinMaxScaler()
ms
```

```python
## fit the minmax model
ms2 = ms.fit_transform(Df)
```

```python
ms2
```

```python
ms2  = pd.DataFrame(ms2, columns = Df.columns)
ms2.head()
```

```python
fig,(ax1,ax2) = plt.subplots(ncols = 2, figsize = (9,5))


ax1.set_title('Original DataSet')
sns.kdeplot(Df['TV'],ax = ax1)
sns.kdeplot(Df['radio'],ax = ax1)
sns.kdeplot(Df['newspaper'],ax = ax1)
sns.kdeplot(Df['sales'],ax = ax1)

ax2.set_title('After Appling MinMax Scaler')
sns.kdeplot(ms2['TV'],ax = ax2)
sns.kdeplot(ms2['radio'],ax = ax2)
sns.kdeplot(ms2['newspaper'],ax = ax2)
sns.kdeplot(ms2['sales'],ax = ax2)
```

# 3.Robust Scaler
- If data set contains larg amount of outliers then this is the best technique.
- Scaling btw inter quartile range(25% to 75%)

- **Formula**
    + [Xi - median(X)] / [Q3(X) - Q1(X)]
        - Q1 - 25%(lower Quartile)
        - Q2 - 50%(median)
        - Q3 - 75%(upper Quartile)

```python
Df.head()
```

```python
Df.tail()
```

```python
Df.quantile(0.25)
```

```python
Df.describe()
```

- **Formula**
    + [Xi - median(X)] / [Q3(X) - Q1(X)]

```python
(38.2 - 149.750000) / (218.825000 - 74.375000)
```

```python
## Import robust model
from sklearn.preprocessing import RobustScaler
```

```python
## Call the robust model and initialise it
rs = RobustScaler()
rs
```

```python
## fit the minmax model
rb = rs.fit_transform(Df)
```

```python
rb
```

```python
rb  = pd.DataFrame(rb, columns = Df.columns)
rb.head()
rb.tail()
```

```python
fig,(ax1,ax2) = plt.subplots(ncols = 2, figsize = (9,5))


ax1.set_title('Original DataSet')
sns.kdeplot(Df['TV'],ax = ax1)
sns.kdeplot(Df['radio'],ax = ax1)
sns.kdeplot(Df['newspaper'],ax = ax1)
sns.kdeplot(Df['sales'],ax = ax1)

ax2.set_title('After Appling Robust Scaler')
sns.kdeplot(rb['TV'],ax = ax2)
sns.kdeplot(rb['radio'],ax = ax2)
sns.kdeplot(rb['newspaper'],ax = ax2)
sns.kdeplot(rb['sales'],ax = ax2)
```
