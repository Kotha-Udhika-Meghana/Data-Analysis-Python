## Catplot

```python
import seaborn as sns
tips_df = sns.load_dataset("tips")
tips_df
```

```python
sns.catplot(x="day",y="total_bill",data=tips_df)
```

```python
help(sns.catplot)
```

```python
## for large dataset we use swarm plot
sns.catplot(x="day",y="total_bill",data=tips_df,kind="swarm")
```

```python
## Jitter is used for only strip plot "kind = strip (default)"
sns.catplot(x="day",y="total_bill",data=tips_df,jitter=False)
```

```python
dir(sns)
```

```python
sns.swarmplot(x="day",y="total_bill",data=tips_df)
```

```python
sns.stripplot(x="day",y="total_bill",data=tips_df)
```

```python
help(sns.swarmplot)
```

```python
sns.catplot(x="smoker",y="tip",data = tips_df)
```

```python
sns.catplot(x="smoker",y="tip",data = tips_df,order=["No","Yes"])
```

```python
sns.catplot(y="day",x="total_bill",data=tips_df,kind="swarm",hue="time")
```

### Catogarical Distribution plots
#### Box Plot

```python
sns.catplot(x="day",y="total_bill",kind="box",data=tips_df,hue="sex")
```

```python
tips_df["weekend"] = tips_df["day"].isin(["Sat","Sun"])
tips_df["weekend"]
```

```python
sns.catplot(x="day",y="total_bill",kind="box",data=tips_df,hue="weekend")
```

```python
sns.catplot(x="day",y="total_bill",kind="box",data=tips_df,
            hue="weekend",dodge=False)
```

#### Boxen Plot

```python
diamonds_df = sns.load_dataset("diamonds")
diamonds_df
```

```python
sns.catplot(x="color",y="price",data=diamonds_df,kind="boxen")
```

```python
sns.catplot(x="color",y="price",data=diamonds_df,kind="box")
```

#### Violin Plot

```python
sns.catplot(x="color",y="price",data=diamonds_df,kind="violin")
```

```python
sns.catplot(y="day",x="total_bill",data=tips_df,kind="violin")
```

```python
sns.catplot(y="day",x="total_bill",data=tips_df,kind="violin",hue="sex")
```

```python
help(sns.violinplot)
```

```python
sns.catplot(y="day",x="total_bill",data=tips_df,
            kind="violin",hue="sex",bw='silverman')
```

```python
sns.catplot(y="day",x="total_bill",data=tips_df,
            kind="violin",hue="sex",bw='silverman',split=True)
```

```python
sns.catplot(x="day",y="total_bill",data=tips_df,
        kind="violin",hue="sex",bw_method='silverman',split=True,inner='stick')
```

```python
a = sns.catplot(x="day",y="total_bill",kind='violin',inner=None,data=tips_df)
sns.swarmplot(x="day",y="total_bill",data=tips_df,ax=a.ax,color='k')
```

### Estimation Plots
- Bar,count,point
### Bar plot

```python
titanic_df=sns.load_dataset("titanic")
titanic_df.head()
```

```python
sns.catplot(x="sex",y="survived",data=titanic_df,kind="bar")
```

```python
sns.catplot(x="sex",y="survived",data=titanic_df,kind="bar",hue="embarked")
```

```python
sns.catplot(x="sex",y="survived",data=titanic_df,kind="bar",hue="class")
```

### Count plot

```python
sns.catplot(x="adult_male",data=titanic_df,kind="count")
```

```python
sns.catplot(x="deck",data=titanic_df,kind="count")
```

```python
sns.catplot(x="who",data=titanic_df,kind="count")
```

```python
sns.catplot(x="who",data=titanic_df,kind="count",hue="class")
```

```python
sns.catplot(x="who",data=titanic_df,kind="count",hue="adult_male")
```

### Point Plot

```python
sns.catplot(x="who",y="adult_male",data=titanic_df,kind="point")
```

```python
sns.catplot(x="sex",y="survived",data=titanic_df,kind="point")
```

```python
sns.catplot(x="age",y="class",data=titanic_df,kind="point")
```

```python
sns.catplot(x="age",y="deck",data=titanic_df,kind="point")
```

```python
sns.catplot(x="sex",y="survived",data=titanic_df,kind="point",hue="class")
```

```python
sns.catplot(x="sex",y="survived",data=titanic_df,kind="point",hue="deck")
```

```python
sns.catplot(x="sex",y="survived",data=titanic_df,kind="point",
            hue="class",palette={"First":'r',"Second":'k',"Third":'g'})
```
