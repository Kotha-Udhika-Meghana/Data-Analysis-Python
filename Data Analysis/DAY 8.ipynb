## Handling Missing Values using Pandas
- Mainly 4 methods are there for identify and manipulate missing values.  
    1.isnull() : Presence of missing values, returns boolean data.  
    2.not null() : opp to isnull().  
    3.dropna() : It removes null values.  
    4.fillna() : Fills the missing values by specified values.

### Replace null values
- If the data is numerical, Replace the Null values with 0 or mean() or median().   
- If the data is categorical, Replace the null value with most occured name or any name.
    * fillna()

```python
df.head(2)
```

```python
data2 = df.copy()
```

```python
data2.head(2)
```

```python
data2.shape
```

```python
data2.isnull().sum().sum()
```

- **Here "Product_Base_Margin"is a Numarical column**

```python
data2['Product_Base_Margin'].mean()
```

```python
data2['Product_Base_Margin'] = data2['Product_Base_Margin'].fillna(data2['Product_Base_Margin'].mean())
```

```python
data2.shape
```

```python
data2.isnull().sum()
```

### ffill and bfill

```python
D2 = {1:[111,112,np.nan,114,115,np.nan],
      2:[221,np.nan,223,224,225,226],
      3:[np.nan,332,np.nan,334,np.nan,336]}
df2 = pd.DataFrame(D2)
```

```python
df2
```

- ffill : front fill
- Appaly ffill for 1st column

```python
df2[1].fillna(method = 'ffill',inplace = True)
```

```python
df2
```

```python
help(df2.fillna)
```

```python
df2[2].fillna(method = 'bfill',inplace = True)
```

```python
df2
```

```python
df2[1].fillna(0,inplace = True)
```

```python
df2
```
