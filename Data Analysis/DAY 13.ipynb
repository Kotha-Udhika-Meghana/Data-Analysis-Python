## 2.Histogram Plot
- plt.hist()

```python
stud_1 = np.random.randint(18,45,100)
stud_2 = np.random.randint(15,40,100)
plt.hist(stud_1,bins=5)
plt.xlabel("Age category")
plt.ylabel("No. of students in that age")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(stud_1,bins,histtype='bar',orientation = 'vertical',color = 'c')
plt.show()
```

```python
help(plt.hist)
```

```python
from matplotlib import style
style.use("ggplot")
plt.figure(figsize = (9,5))
plt.hist([stud_1,stud_2],bins,rwidth = 0.8,histtype = 'bar',orientation = 'vertical',color = ['m','y'])
plt.title("Hiatogram Example")
plt.xlabel("Age category")
plt.ylabel("No. of students in that age")
plt.show()
```

## 3.Pie plot
- plt.pie()

```python
a = [15,25,25,30,5]
plt.pie(a)
plt.show()
```

### Labels and values for pie chart

```python
labels = ['Very Likely','Likely','Unsure','Unlikely','Very Unlikely']
plt.pie(a,labels = labels)
plt.show()
```

### Adding colours

```python
colors = ['magenta','green','red','yellow','cyan']
plt.pie(a,labels = labels,colors = colors)
plt.show()
```

### Adding Values

```python
plt.pie(a,labels = labels,colors = colors,autopct = "%.0f%%")
plt.title("Survey Responses")
plt.show()
```

### Excluding/Removing parts in pie plot

```python
explode = [0,0,0,0,0.5]
plt.pie(a,labels = labels,colors = colors,autopct = "%.0f%%",explode = explode)
plt.title("Survey Responses")
plt.show()
```

### Adding shows pie chart

```python
explode = [0,0,0,0,0.5]
plt.pie(a,labels = labels,colors = colors,autopct = "%.0f%%",explode = explode,shadow='True')
plt.title("Survey Responses")
plt.show()
```

### Adding Angles

```python
explode = [0,0,0,0,0.5]
plt.pie(a,labels = labels,colors = colors,autopct = "%.0f%%",explode = explode,shadow='True',startangle=180)
plt.title("Survey Responses")
plt.show()
```

## 4. Stack Plot
- plt.stackplot()

```python
p = [1,2,3,4,5]
q1 = [5,6,4,5,7]
q2 = [1,5,6,4,6]
q3 = [1,1,2,3,2]
plt.stackplot(p,q1,q2,q3)
plt.show()
```

```python
p = [1,2,3,4,5]
q_values = {"q1" : [5,6,4,5,7],"q2" : [1,5,6,4,6],"q3" : [1,1,2,3,2]}
plt.stackplot(p,q_values.values())
plt.show()
```

```python
plt.stackplot(p,q_values.values(),labels = q_values.keys())
plt.legend(loc="upper left")
plt.show()
```

## 5.Box plot
- plt.boxplot()

```python
data1 = pd.read_csv("winequality-red.csv")
data1.head()
```

```python
data1.isnull().sum().sum()
```

```python
data1.isnull().values.any()
```

```python
fixed_acid = data1["fixed acidity"]
free_sulphur = data1["free sulfur dioxide"]
total_sulphur = data1["total sulfur dioxide"]
plt.boxplot(fixed_acid)
plt.show()
```

```python
plt.boxplot(fixed_acid,vert=0)
plt.show()
```

```python
plt.boxplot([fixed_acid,free_sulphur,total_sulphur])
plt.show()
```

```python
plt.boxplot([fixed_acid,free_sulphur,total_sulphur],notch = True)
plt.show()
```

```python
plt.boxplot([fixed_acid,free_sulphur,total_sulphur],notch = True,patch_artist = True)
plt.show()
```

```python
plt.boxplot([fixed_acid,free_sulphur,total_sulphur],notch = True,patch_artist = True,showmeans = True)
plt.show()
```
